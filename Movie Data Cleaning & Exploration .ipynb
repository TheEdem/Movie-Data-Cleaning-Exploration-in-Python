```python
# Import all necessary packages for this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


#read in dataset
df= pd.read_csv(r'C:\Users\there\Downloads\movies.csv')
```

```python
#overview of dataset
df.head()
```

```python
# Check for any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Check for any Outliers?

df.boxplot(column=['gross'])
```

```python
#drop duplicates
df.drop_duplicates()
```

```python
# Prompt python to display full table
# Order Data 
pd.set_option('display.max_rows', none)
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#changing column data types
df['budget'].round().astype('float')
df['budget']=df['budget'].astype('float')
df['gross'].round().astype('float')
df['gross']=df['gross'].astype('float')
df
```

```python
#Creating a corrected year column for dataset
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
df
```

```python
#Checking for correlations


#scatter plot of budget vs gross earnings

plt.scatter(x=df['budget'], y=df['gross'])
```

```python
#create regression plot for budget vs gross using seaborn
sns.regplot(x="gross", y="budget", data=df)
```

```python
#Checking other correlation methods
df.corr(method = 'pearson', numeric_only = True)
```

```python
#Checking other correlation methods
df.corr(method = 'kendall', numeric_only = True)
```

```python
#Checking other correlation methods
df.corr(method = 'spearman', numeric_only = True)
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
